### Get data

```python
import pandas as pd
import sqlalchemy as sa
import requests
```

```python
resource_id = '15b54ed9-f711-45c8-bd87-30eb626f7c0a'
limit = 10500
offset = 0

api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)

print(api_url)
```

```python
api_response = requests.get(api_url)
api_response
```

```python
data = api_response.json()
data
```

```python
data['result']['records']
```

```python
ontario_oil_price = pd.DataFrame(data['result']['records'])
ontario_oil_price
```

### Data Engineering

```python
ontario_oil_price['reported_date'] = pd.to_datetime(ontario_oil_price['Date'])
ontario_oil_price['Toronto_West'] = ontario_oil_price['Toronto West/Ouest']
ontario_oil_price['Toronto_West_Difference'] = ontario_oil_price['Toronto_West'].diff(1)
ontario_oil_price['Toronto_East'] = ontario_oil_price['Toronto East/Est']
ontario_oil_price['Toronto_East_Difference'] = ontario_oil_price['Toronto_East'].diff(1)
ontario_oil_price['Toronto_West_Difference'].fillna(0,inplace = True)
ontario_oil_price['Toronto_East_Difference'].fillna(0,inplace = True)

relevant_columns = ['reported_date', 'Toronto_West','Toronto_West_Difference', 'Toronto_East', 'Toronto_East_Difference']

toronto_oil_price_and_difference = ontario_oil_price[relevant_columns]
toronto_oil_price_and_difference
```
