```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = pd.read_csv("/Users/justinsheng/Library/CloudStorage/OneDrive-UniversityofToronto/Justin/Schulich_2023-2024/MMAI5100/Group_Project/clean_data.csv")
```

```python
data.info()
```

```python
X = data.drop(columns=["HURT"]) 
y = data["HURT"] 
```

```python
X = pd.get_dummies(X, columns=["OCC_MONTH", "OCC_DOW", "OCC_TIME_RANGE", "NEIGHBOURHOOD_158"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1001)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=1001)
```

```python
rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print("模型评估结果：")
print("准确率：", accuracy)
print("精确率：", precision)
print("召回率：", recall)
print("F1 分数：", f1)
print("混淆矩阵：")
print(conf_matrix)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.formula.api import ols
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
import seaborn as sns
```

```python
crime = pd.read_csv('/Users/justinsheng/Library/CloudStorage/OneDrive-UniversityofToronto/Justin/Schulich_2023-2024/MMAI5100/Group_Project/clean_data.csv')
```

```python
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July' : 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}


crime['OCC_MONTH'] = crime['OCC_MONTH'].map(month_mapping)
```

```python
date_mapping = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}


crime['OCC_DOW'] = crime['OCC_DOW'].map(date_mapping)
```

```python
results = ols("HURT ~ OCC_YEAR+OCC_DOY+OCC_HOUR+HOOD_158", data=crime).fit()
print(results.summary())
```

```python
results_1 = ols("HURT ~ OCC_YEAR+OCC_HOUR+NEIGHBOURHOOD_158", data=crime).fit()
print(results_1.summary())
```

```python
categorical_features = ['NEIGHBOURHOOD_158']
numerical_features = ['OCC_YEAR','OCC_DOY','OCC_HOUR']
target = 'HURT'
```

```python
X_train, X_test, y_train, y_test = train_test_split(crime[categorical_features + numerical_features]
                                                    ,crime[target],
                                                    test_size = 0.3, random_state=1234)
```

```python
cat_transformer = Pipeline(steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))])
num_transformer = Pipeline(steps = [('scaler', StandardScaler())])
```

```python
preprocessor = ColumnTransformer(transformers = [('cat',cat_transformer, categorical_features),
                                                 ('num', num_transformer, numerical_features)])
```

```python
linear_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
linear_regression.fit(X_train, y_train)
```

```python
pred = linear_regression.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(pred,y_test))
```

```python
ridge_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=3.0))
])
```

```python
ridge_regression.fit(X_train, y_train)
```

```python
pred_ridge = ridge_regression.predict(X_test)
```

```python
pred_ridge
```

```python

np.sqrt(mean_squared_error(pred_ridge,y_test))
```

```python
lasso_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=1.0))
])
```

```python
lasso_regression.fit(X_train, y_train)
```

```python
lasso_pred = lasso_regression.predict(X_test)
```

```python
lasso_pred
```

```python
np.sqrt(mean_squared_error(lasso_pred, y_test))
```

```python
elasticnet_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', ElasticNet(alpha=1.0, l1_ratio=0.8))
])
```

```python

elasticnet_regression.fit(X_train, y_train)
```

```python
elasticnet_pred = elasticnet_regression.predict(X_test)
```

```python
elasticnet_pred
```

```python

np.sqrt(mean_squared_error(elasticnet_pred, y_test))
```

```python
coefficients = linear_regression.named_steps['regressor'].coef_
feature_names = list(crime[numerical_features].columns)
print("Coefficients:")
for feature, coef in zip(feature_names, coefficients):
    print(f"{feature}: {coef}")
    
print(f"Intercept: {elasticnet_regression.named_steps['regressor'].intercept_}")
```
